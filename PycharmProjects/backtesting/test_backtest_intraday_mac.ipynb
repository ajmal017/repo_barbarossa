```python
import backtesting.backtest_intraday_mac as bim
trend_detection_period = 50
trigger_period = 4
stop_multiplier = 1
target_multiplier = 2.5
ticker_head = 'LC'
results_out = bim.get_results_4given_inputs(ticker_head=ticker_head,
                        trend_detection_period=trend_detection_period,trigger_period=trigger_period,
                        stop_multiplier = stop_multiplier,target_multiplier=target_multiplier)
```

```python
trades_frame = results_out['trades_frame']
trades_frame['pnl_dollar'].sum()
```

```python
import matplotlib.pyplot as plt
plt.plot(range(len(trades_frame.index)),trades_frame['pnl_dollar'].cumsum())
plt.grid()         
         
plt.show()
```

```python
import matplotlib.pyplot as plt
candle_frame = results_out['candle_frame']
start = 760
end = 860
select_frame = candle_frame[start:end]
index = range(start,end)
plt.plot(index,select_frame['close'],
         index,select_frame['trigger_ma'],
         index,select_frame['trend_ma'])
plt.grid()         
         
plt.show()
```
