```python
import sys
sys.path.append(r'C:\Users\kocat_000\quantFinance\PycharmProjects')
import backtesting.utilities as bu
dates_output = bu.get_backtesting_dates(date_to=20151228,years_back=10)
double_dates = sorted(dates_output['double_dates'],reverse=True)
selected_dates = double_dates
selected_dates
```

```python

report_results_list = []

import opportunity_constructs.curve_pca as cpc

for date_to in selected_dates:
    print(date_to)
    report_out = cpc.get_curve_pca_report(ticker_head='CL',date_to=date_to,use_existing_filesQ=False)
    report_results_list.append(report_out['pca_results'])

```

```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

total_pnl_list = []

for i in range(len(selected_dates)):
    daily_report = report_results_list[i]
    
    median_factor_load2 = daily_report['factor_load2'].median()
    
    if median_factor_load2>0:
        daily_report_filtered = daily_report[daily_report['factor_load2']>=0]
    else:
        daily_report_filtered = daily_report[daily_report['factor_load2']<=0]
    
    daily_report_filtered.sort('z',ascending=True,inplace=True)
    num_contract_4side = round(len(daily_report_filtered.index)/4)
    long_side = daily_report_filtered.iloc[:num_contract_4side]
    short_side = daily_report_filtered.iloc[-num_contract_4side:]
    short_side_weight = long_side['factor_load1'].sum()/short_side['factor_load1'].sum()
    total_pnl_list.append(long_side['change5'].mean()-short_side_weight*short_side['change5'].mean())
```

```python
import numpy as np
pnl_frame = pd.DataFrame({'pnl':total_pnl_list,'settle_date':selected_dates })
pnl_frame_sorted = pnl_frame.sort('settle_date',ascending=True,inplace=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(pnl_frame_sorted['pnl'].cumsum())
plt.grid()               
plt.show()
```
