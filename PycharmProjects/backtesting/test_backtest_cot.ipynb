```python
# Portfolio Construction:

# Seem to have some relevance for FX

import pandas as pd
combined_frame = pd.concat(combined_data_list)

weekly_frame = combined_frame[combined_frame['spec_long'].notnull()]
weekly_frame = weekly_frame[weekly_frame['ticker_head'].isin(['JY','EC','AD','CD','BP'])]
#weekly_frame = weekly_frame[weekly_frame['ticker_head'].isin(['LC','LN','GF'])]
#weekly_frame = weekly_frame[weekly_frame['ticker_head'].isin(['C','W','S','SM','BO','KW'])]
settle_date_list = weekly_frame['settle_date'].unique()
long_pnl_list = []
short_pnl_list = []

for i in range(len(settle_date_list)):
    data_4week = weekly_frame[weekly_frame['settle_date']==settle_date_list[i]]
    
    
    weekly_long = data_4week[(data_4week['spec_long']>data_4week['spec_long_ma13'])&
                                (data_4week['spec_short']<data_4week['spec_short_ma13'])&
                                (data_4week['spec_long_per']>30)]
    
    weekly_short = data_4week[(data_4week['spec_long']<data_4week['spec_long_ma13'])&
                                (data_4week['spec_short']>data_4week['spec_short_ma13'])&
                                (data_4week['spec_short_per']>60)]
    
    long_pnl_list.append(weekly_long['const_mat_change5norm'].sum())
    short_pnl_list.append(-weekly_short['const_mat_change5norm'].sum())
    
    
pnl_frame = pd.DataFrame.from_items([('settle_date', settle_date_list), ('long', long_pnl_list),
                        ('short', short_pnl_list)])

pnl_frame['total'] = pnl_frame['long'] + pnl_frame['short'] 
```

```python
date_list = backtest_frame['settle_date'].unique()
import numpy as np
pnl_list = []

for i in range(len(date_list)):
    
    daily_data = backtest_frame[backtest_frame['settle_date']==date_list[i]]
    daily_data.sort('change_40_normalized',ascending=True,inplace=True)
    short_pnl = -np.sum(daily_data['normalized_pnl_5'].iloc[:3])
    long_pnl = np.sum(daily_data['normalized_pnl_5'].iloc[-3:])
    pnl_list.append(short_pnl+long_pnl)
```

```python


import matplotlib.pyplot as plt

plt.plot(np.cumsum(pnl_list))
plt.grid()         
         
plt.show()
```
