```python
# Tickerhead Results:

big_data = wuhu['big_data']
import opportunity_constructs.spread_carry as sc
import backtesting.backtest_stats as bs
import pandas as pd
ticker_head_list = list(sc.max_tr_dte_limits.keys())

long_stats = [bs.get_summary_stats(big_data.loc[(big_data['q_carry']>5)&
                                  (big_data['ticker_head']==x)
                                  ,'pnl_final'].values)
                                  for x in ticker_head_list]
short_stats = [bs.get_summary_stats(big_data.loc[(big_data['q_carry']<-5)&
                                  (big_data['ticker_head']==x)
                                  ,'pnl_final'].values)
                                  for x in ticker_head_list]

pd.DataFrame.from_items([('tickerHead',ticker_head_list ),
                         ('reward_risk_long', [x['reward_risk'] for x in long_stats]),
                         ('reward_risk_short', [x['reward_risk'] for x in short_stats])])

```

```python
import backtesting.utilities as bu
num_buckets = 10
import backtesting.backtest_stats as bs
import pandas as pd
import shared.statistics as stats
import numpy as np
indicator = 'q_carry'

bucket_limits = bu.get_equal_length_bucket_limits(min_value=0,
                                                  max_value=100,
                                                  num_buckets=num_buckets)

bucket_limits = stats.get_number_from_quantile(y=big_data[indicator].values,quantile_list=bucket_limits)

mean_pnl_list = []
mean_indicator_list = []
reward_risk_list = []

for i in range(num_buckets):
    
    if i==0:
        bucket_data = big_data.loc[big_data[indicator]<=bucket_limits[i],['pnl_long','pnl_short',indicator]]
    elif i<num_buckets-1:
        bucket_data = big_data.loc[(big_data[indicator]>bucket_limits[i-1])&
                                   (big_data[indicator]<=bucket_limits[i]),['pnl_long','pnl_short',indicator]]
    else:
        bucket_data = big_data.loc[big_data[indicator]>bucket_limits[i-1],['pnl_long','pnl_short',indicator]]
    
    if i<=(num_buckets/2):
        signed_pnl = bucket_data['pnl_short']
    else:
        signed_pnl = bucket_data['pnl_long']
        
    stats_output = bs.get_summary_stats(signed_pnl.values)    
    mean_pnl_list.append(stats_output['mean_pnl'])
    mean_indicator_list.append(bucket_data[indicator].mean())
    reward_risk_list.append(stats_output['reward_risk'])
        

pd.DataFrame.from_items([('indicator_upper_limit', np.append(bucket_limits,np.NAN)),
                         ('mean_pnl', mean_pnl_list),
                         ('reward_risk',reward_risk_list)])
```
