```python
import backtesting.backtest_futures_directional as bfd

backtest_output_list = []
corr_matrix_list = []


for i in range(len(date_list)):
    print(date_list[i])
    directionals_output = bfd.get_results_4date(date_to=date_list[i])
    directionals = directionals_output['directionals']
    directionals['report_date'] = date_list[i]
    backtest_output_list.append(directionals)
    corr_matrix_list.append(directionals_output['corr_matrix'])
```

```python
import portfolioopt as pfopt

long_frame = backtest_frame[backtest_frame['regress_forecast3']>0.07]
short_frame = backtest_frame[backtest_frame['regress_forecast3']<-0.07]

long_frame['pnl'] = long_frame['norm_pnl1Instant']
short_frame['pnl'] = short_frame['norm_pnl1Instant_']

merged_frame  = pd.concat([long_frame,short_frame])
merged_frame['pnl'].mean()/merged_frame['pnl'].std()
report_date_list = merged_frame['report_date'].unique()

weighted_long_pnl_list = []
weighted_short_pnl_list = []
long_pnl_list = []
short_pnl_list = []

for i in range(0,len(report_date_list)):
    print(report_date_list[i])
    corr_matrix = corr_matrix_list[i]
    daily_frame = merged_frame[merged_frame['report_date']==report_date_list[i]]
    long_frame = daily_frame[(daily_frame['regress_forecast2']>0.06)&(daily_frame['regress_forecast1']>0.07)]
    short_frame = daily_frame[(daily_frame['regress_forecast2']<-0.06)&(daily_frame['regress_forecast1']<-0.07)]
    
    weighted_long_pnl = 0
    weighted_short_pnl = 0
    long_pnl = 0
    short_pnl = 0
    

    if not long_frame.empty:
        long_pnl = long_frame['long_loose_stop_pnl1Instant'].sum()
        if len(set(long_frame['ticker_head'].values)-set(corr_matrix.columns))==0:
            long_corr_matrix = corr_matrix.loc[long_frame['ticker_head'].values,long_frame['ticker_head'].values]
            if not long_corr_matrix.isnull().any().any():
                weights = len(long_frame.index)*pfopt.min_var_portfolio(long_corr_matrix)
                weighted_long_pnl = sum(long_frame['norm_pnl1Instant'].values*weights)
    
    weighted_long_pnl_list.append(weighted_long_pnl)
    long_pnl_list.append(long_pnl)

    if not short_frame.empty:
        short_pnl = short_frame['short_loose_stop_pnl1Instant'].sum()
        if len(set(short_frame['ticker_head'].values)-set(corr_matrix.columns))==0:
            short_corr_matrix = corr_matrix.loc[short_frame['ticker_head'].values,short_frame['ticker_head'].values]
            if not short_corr_matrix.isnull().any().any():
                weights = len(short_frame.index)*pfopt.min_var_portfolio(short_corr_matrix)
                weighted_short_pnl = sum(short_frame['norm_pnl1Instant_'].values*weights)    

    weighted_short_pnl_list.append(weighted_short_pnl)
    short_pnl_list.append(short_pnl)
                                           
                                        
portfolio_frame = pd.DataFrame.from_items([('report_date', report_date_list),
                                           ('wlong', weighted_long_pnl_list),
                                           ('wshort', weighted_short_pnl_list),
                                           ('long', long_pnl_list),
                                           ('short', short_pnl_list)])
```

```python
portfolio_frame['wtotal'] = portfolio_frame['wlong']+portfolio_frame['wshort']
portfolio_frame['total'] = portfolio_frame['long']+portfolio_frame['short']
```

```python
portfolio_frame

import matplotlib.pyplot as plt

plt.plot(range(len(portfolio_frame.index)),portfolio_frame['wtotal'].cumsum(),
         range(len(portfolio_frame.index)),portfolio_frame['total'].cumsum())
plt.grid()   
plt.legend(['wtotal','total'])
         
plt.show()
```

```python


import shared.statistics as stats
import pandas as pd

indicator_list = ['change_1_high_volume','change_5_high_volume','change_10_high_volume','change_20_high_volume',
                      'change_1_low_volume','change_5_low_volume','change_10_low_volume','change_20_low_volume',
                      'change_1_norm','change_5_norm','change_10_norm','change_20_norm',
                      'comm_net_change_1_normalized','comm_net_change_2_normalized','comm_net_change_4_normalized',
                      'spec_net_change_1_normalized','spec_net_change_2_normalized','spec_net_change_4_normalized',
                      'comm_z','spec_z','vote1','vote1_instant','vote12_instant','vote13_instant','vote5','vote10','vote20']
sharp1_instant_list = []
sharp1_list = []
sharp5_list = []
sharp10_list = []
sharp20_list = []

selected_frame = backtest_frame[backtest_frame['weekday']==0]
selected_frame = backtest_frame
selected_frame = backtest_frame[backtest_frame['ticker_head']=='US']

for i in range(len(indicator_list)):
    indicator_levels = stats.get_number_from_quantile(y=selected_frame[indicator_list[i]].values,quantile_list=[10,90])
    high_frame = selected_frame[selected_frame[indicator_list[i]]>=indicator_levels[1]]
    low_frame = selected_frame[selected_frame[indicator_list[i]]<=indicator_levels[0]]
    low_frame['norm_pnl1Instant'] = -low_frame['norm_pnl1Instant']
    low_frame['norm_pnl1'] = -low_frame['norm_pnl1']
    low_frame['norm_pnl5'] = -low_frame['norm_pnl5']
    low_frame['norm_pnl10'] = -low_frame['norm_pnl10']
    low_frame['norm_pnl20'] = -low_frame['norm_pnl20']
    merged_data = pd.concat([high_frame,low_frame])    
    sharp1_instant_list.append(16*merged_data['norm_pnl1Instant'].mean()/merged_data['norm_pnl1Instant'].std())
    sharp1_list.append(16*merged_data['norm_pnl1'].mean()/merged_data['norm_pnl1'].std())
    sharp5_list.append(7.2*merged_data['norm_pnl5'].mean()/merged_data['norm_pnl5'].std())
    sharp10_list.append(5.1*merged_data['norm_pnl10'].mean()/merged_data['norm_pnl10'].std())
    sharp20_list.append(3.5*merged_data['norm_pnl20'].mean()/merged_data['norm_pnl20'].std())
    
sharp_frame = pd.DataFrame.from_items([('indicator', indicator_list),
                         ('sharp1_instant', sharp1_instant_list),
                         ('sharp1', sharp1_list),
                         ('sharp5', sharp5_list),
                         ('sharp10', sharp10_list),
                         ('sharp20', sharp20_list)])

sharp_frame
```

```python
# sharp ratios for vote1 indicator

report_date_list = backtest_frame['report_date'].unique()
long_pnl_list = []
short_pnl_list = []

for i in range(len(report_date_list)):
    daily_frame = backtest_frame[backtest_frame['report_date']==report_date_list[i]]
    long_frame = daily_frame[daily_frame['vote20']>5]
    short_frame = daily_frame[daily_frame['vote20']<-5]
    
    if long_frame.empty:
        long_pnl_list.append(0)
    else:
        long_pnl_list.append(long_frame['norm_pnl5'].sum())
    
    if short_frame.empty:
        short_pnl_list.append(0)
    else:
        short_pnl_list.append(-short_frame['norm_pnl5'].sum())
        
    
    
portfolio_frame = pd.DataFrame.from_items([('report_date', report_date_list),
                                           ('long_pnl', long_pnl_list), 
                                           ('short_pnl', short_pnl_list)])    

portfolio_frame['total_pnl'] = portfolio_frame['long_pnl']+portfolio_frame['short_pnl']

print('Total sharp: ' +  str(16*portfolio_frame['total_pnl'].mean()/ portfolio_frame['total_pnl'].std()))
print('Long sharp: ' +  str(16*portfolio_frame['long_pnl'].mean()/ portfolio_frame['long_pnl'].std()))
print('Short sharp: ' +  str(16*portfolio_frame['short_pnl'].mean()/ portfolio_frame['short_pnl'].std()))

# >=1 gives 1.39 sharp ratio
# >=2 gives 1.55 sharp ratio
# >=3 gives 2.4 sharp ratio
# >=4 gives 1.6 sharp ratio

```
