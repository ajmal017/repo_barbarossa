```python
import backtesting.backtest_stats as bs
import pandas as pd
from IPython.display import display
import backtesting.backtest_stats as bs
bs.get_indicator_rr_double_table(trade_data=backtest_frame,indicator_list=['regress_forecast52', 'regress_forecast54'],
                                 strategy_class='ofs',
                                 long_pnl_field='change5Normalized',
                                            short_pnl_field='change5Normalized_')
```

```python
# Portfolio Sharp Ratio:

# Instant 1 day sharp ratio: 0.87
# 5 day sharp ratio: 0.34

report_date_list = backtest_frame['report_date'].unique()

long_pnl_list = []
short_pnl_list = []

for i in range(len(report_date_list)):
    selected_frame = backtest_frame[backtest_frame['report_date']==report_date_list[i]]
    long_frame = selected_frame[(selected_frame['regress_forecast52']>0.1)&(selected_frame['regress_forecast54']>0.35)]
    short_frame = selected_frame[(selected_frame['regress_forecast52']<-0.1)&(selected_frame['regress_forecast54']<-0.35)]
    
    if not long_frame.empty:
        long_pnl_list.append(long_frame['change5Normalized'].sum())
    else:
        long_pnl_list.append(0)
    
    if not short_frame.empty:
        short_pnl_list.append(short_frame['change5Normalized_'].sum())
    else:
        short_pnl_list.append(0)


portfolio_frame = pd.DataFrame.from_items([('report_date', report_date_list),
                                           ('long', long_pnl_list),
                                           ('short', short_pnl_list)])        
        
    
portfolio_frame['total'] = portfolio_frame['long']+portfolio_frame['short']

print(7.2*portfolio_frame['total'].mean()/portfolio_frame['total'].std())
```

```python
spread_description_list = backtest_frame['spread_description'].unique()

long_pnl_list = []
short_pnl_list = []

for i in range(len(spread_description_list)):
    selected_frame = backtest_frame[backtest_frame['spread_description']==spread_description_list[i]]
    long_frame = selected_frame[(selected_frame['regress_forecast1Instant2']>0.03)&(selected_frame['regress_forecast1Instant4']>0.1)]
    short_frame = selected_frame[(selected_frame['regress_forecast1Instant2']<-0.03)&(selected_frame['regress_forecast1Instant4']<-0.1)]
    
    if not long_frame.empty:
        long_pnl_list.append(16*long_frame['change1_instantNormalized'].mean()/long_frame['change1_instantNormalized'].std())
    else:
        long_pnl_list.append(0)
    
    if not short_frame.empty:
        short_pnl_list.append(16*short_frame['change1_instantNormalized_'].mean()/short_frame['change1_instantNormalized'].std())
    else:
        short_pnl_list.append(0)


portfolio_frame = pd.DataFrame.from_items([('spread_description', spread_description_list),
                                           ('long', long_pnl_list),
                                           ('short', short_pnl_list)])       

portfolio_frame
```

```python


import matplotlib.pyplot as plt

plt.plot(range(len(portfolio_frame.index)),portfolio_frame['total'].cumsum())
plt.grid()   
         
plt.show()
```
