```python

import opportunity_constructs.overnight_calendar_spreads as ocs
import pandas as pd

backtest_result_list = []

for i in range(len(date_list)):
    print(date_list[i])
    backtest_output = ocs.generate_overnight_spreads_sheet_4date(date_to=date_list[i])
    if backtest_output['success']:
        overnight_calendars = backtest_output['overnight_calendars']
        overnight_calendars['report_date'] = date_list[i]
        backtest_result_list.append(backtest_output['overnight_calendars'])
        
backtest_frame = pd.concat(backtest_result_list)
backtest_frame = backtest_frame[backtest_frame['trDte1']>0]
```

```python
# Clean Extreme Moves Here But Will Need To Check For These Later

backtest_frame = backtest_frame[(backtest_frame['normalized_target']>-5)&(backtest_frame['normalized_target']<5)]
backtest_frame['normalized_target_'] = -backtest_frame['normalized_target']
backtest_frame['normalized_target_c_'] = -backtest_frame['normalized_target_c']
```
