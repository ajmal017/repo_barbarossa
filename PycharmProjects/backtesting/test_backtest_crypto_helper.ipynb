```python
# Get Intraday Data
import get_price.get_binance_price as gbp
output = gbp.get_binance_price_preloaded(ticker='LINKBTC',interval='4H',date_from=20170714,date_to=20200424)
output['change_1'] = output['close'].diff()
```

```python
import signals.technical_indicators as ti
ticker_data = ti.rsi(data_frame_input=output[~output['change_1'].isnull()],change_field='change_1',period=14)
ticker_data['ma300'] = ticker_data['close'].rolling(window=300, center=False, min_periods=250).mean()
ticker_data['ma30'] = ticker_data['close'].rolling(window=30, center=False, min_periods=25).mean()

ticker_data['trend_filter'] = False
ticker_data.loc[ticker_data['ma30'] > ticker_data['ma300'], 'trend_filter'] = True

ticker_data
```
