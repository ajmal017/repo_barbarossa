```python
import opportunity_constructs.utilities as ocu
import shared.statistics as stats
data_frame = ocu.load_aligend_options_data_file(ticker_head='US',
                                                tr_dte_center=50)  #50
#selected = wuhu[wuhu['settleDates']==20150223]


data_frame = data_frame[(data_frame['settleDates']>20100223)&(data_frame['settleDates']<=20150223)]

profit5 = data_frame['profit5']

percentile_vector = stats.get_number_from_quantile(y=profit5.values,
                                                       quantile_list=[1, 15, 85, 99],
                                                       clean_num_obs=max(100, round(3*len(profit5.values)/4)))

print('downside: ' + str((percentile_vector[0]+percentile_vector[1])/2))
print('upside: ' + str((percentile_vector[2]+percentile_vector[3])/2))

print('tickerYear: ' + str(data_frame['TickerYear'].iloc[-1]))
print('tickerMonth: ' + str(data_frame['TickerMonth'].iloc[-1]))
print('trDTE: ' + str(data_frame['trDTE'].iloc[-1]))

print('dollarTheta: ' + str(data_frame['dollarTheta'].iloc[-1]))

print('normTheta: ' + str(data_frame['normDollarTheta'].iloc[-1]))



```

```python
import pandas as pd

data_frame1 = ocu.load_aligend_options_data_file(ticker_head='CB',contract_month_letter='U',
                                                tr_dte_center=140)

data_frame1 = data_frame1[(data_frame1['settleDates']>20100223)&(data_frame1['settleDates']<=20150223)]

data_frame1

data_frame2 = ocu.load_aligend_options_data_file(ticker_head='CB',contract_month_letter='Z',
                                                tr_dte_center=200)

data_frame2 = data_frame2[(data_frame2['settleDates']>20100223)&(data_frame2['settleDates']<=20150223)]

data_frame2


merged_data = pd.merge(data_frame1,data_frame2,how='inner',on='settleDates')

profit5 = merged_data['profit5_x']-merged_data['profit5_y']

percentile_vector = stats.get_number_from_quantile(y=profit5.values,
                                                       quantile_list=[1, 15, 85, 99],
                                                       clean_num_obs=max(100, round(3*len(profit5.values)/4)))

print('downside: ' + str((percentile_vector[0]+percentile_vector[1])/2))
print('upside: ' + str((percentile_vector[2]+percentile_vector[3])/2))


print('tickerYear1: ' + str(merged_data['TickerYear_x'].iloc[-1]))
print('tickerMonth1: ' + str(merged_data['TickerMonth_x'].iloc[-1]))
print('trDTE1: ' + str(merged_data['trDTE_x'].iloc[-1]))

print('tickerYear2: ' + str(merged_data['TickerYear_y'].iloc[-1]))
print('tickerMonth2: ' + str(merged_data['TickerMonth_y'].iloc[-1]))
print('trDTE2: ' + str(merged_data['trDTE_y'].iloc[-1]))

print('dollarTheta: ' + str(merged_data['dollarTheta_x'].iloc[-1]-merged_data['dollarTheta_y'].iloc[-1]))


```
